### Scraping Rotten Tomatoes

```python
movie=[]
for url in urls:
  responce=requests.get(url)
  soup=BeautifulSoup(responce.content,'lxml')
  # movie name
  title=soup.title.contents[0]
  # movie Time,Year
  movie_info=soup.find("div",class_="thumbnail-scoreboard-wrap").find('score-board').p.contents[0].split(',')
  year=int(movie_info[0])
  movie_time=movie_info[-1]
  # movie image
  image=soup.find('div',class_='movie-thumbnail-wrap').find('div').find('img').get('src')
  # movie scores
  movie_score=soup.find("div",class_="thumbnail-scoreboard-wrap").find('score-board')
  audiencescore=int(movie_score.get('audiencescore'))
  tomatometerscore=int(movie_score.get('tomatometerscore'))

  reviews=movie_score.find_all('a')
  tomatometer_score_reviews=int(reviews[0].contents[0].split()[0])
  audience_score_reviews=reviews[1].contents[0]
  # some information about movie
  movie_info=soup.find('div',class_="panel-body content_body")
  movie_information=movie_info.find('div',class_="movie_synopsis clamp clamp-6 js-clamp").contents[0].strip()
  more_info=movie_info.find('ul').find_all('div')
  # movie Genre
  movie_Genre=more_info[1].contents[0].strip()
  # movie language
  movie_Language=more_info[3].contents[0].strip()
  # movie Directors
  movie_Director=more_info[5].find_all('a')
  movie_Directors='/'.join([x.contents[0].strip() for x in movie_Director])
  # movie Producers
  movie_Producer=more_info[7].find_all('a')
  movie_Producers='/'.join([x.contents[0].strip() for x in movie_Producer])
  # movie Writers
  movie_Writer=more_info[9].find_all('a')
  movie_Writers='/'.join([x.contents[0].strip() for x in movie_Writer])
  # movie Date
  # Release_Date_Theaters=more_info[11].time.contents[0].strip()
  # Release_Date_Streaming=more_info[13].time.contents[0].strip()

  # movie cast and role of each one in moive
  cast=soup.find('div',class_="castSection").find_all('span')
  casts=[]
  cast_role=[]
  for x in cast:

    y=x.contents[0].strip()
    if y!='':
      casts.append(y)
    else:
      y=x.contents[2].strip()
      cast_role.append(y)
            
  movie_cast='/'.join(casts)
  cast_role ='/'.join(cast_role)
  movie.append({
          'Title':title,
          'Year':year,
          'Image':image,
          'Movie_Time':movie_time,
          'AudienceScore':audiencescore,
          'TomatometerScore':tomatometerscore,
          'Audience_Score_Reviews':audience_score_reviews,
          'Tomatometer_Score_Reviews':tomatometer_score_reviews,
          'Info':movie_information,
          'Genre':movie_Genre,
          'Language':movie_Language,
          'Directors':movie_Directors,
          'Producers':movie_Producers,
          'Writers':movie_Writers,
          # 'Release_Date_Theaters':Release_Date_Theaters,
          # 'Release_Date_Streaming':Release_Date_Streaming,
          'Cast':movie_cast,
          'Cast_Role':cast_role,
            })

    # except Exception as e:
    #         print(str(e))

df=pd.DataFrame(movie)
```

Extract Data From (*.txt) file in python


```python

df_list=[]
for file in glob.glob(folder_name+'/*.txt'):
    with open(file,encoding="utf_8") as file:
        movie_title=file.readline()[:-1]
        review_url=file.readline()[:-1]
        review_text=file.readlines()
        df_list.append({"movie_title":movie_title,
                        "review_url":review_url,
                        "review_text":review_text,
                       })  
df2=pd.DataFrame(data=df_list,columns=["movie_title","review_url","review_text"])
print(df2)

df2.to_csv("movies_reviews",index=False)
```
